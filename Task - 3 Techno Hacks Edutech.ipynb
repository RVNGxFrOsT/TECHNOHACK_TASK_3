```python
dataset = pd.read_csv("Iris.csv")
Id = dataset["Id"]
SepalLengthCm = dataset["SepalLengthCm"]
SepalWidthCm = dataset["SepalWidthCm"]
PetalLengthCm = dataset["PetalLengthCm"]
PetalWidthCm = dataset["PetalWidthCm"]
dataset.head()
```

# Plotting Histogram

```python
plt.hist(SepalLengthCm, bins = 10, alpha = 0.9, rwidth = 0.8, color = "orange", label = "SepalLengthCm")
plt.hist(SepalWidthCm, bins = 10, alpha = 0.9, rwidth = 0.8, color = "green", label = "SepalLengthCm")
plt.hist(PetalLengthCm, bins = 10, alpha = 0.9, rwidth = 0.8, color = "blue", label = "PetalLengthCm")
plt.hist(PetalWidthCm, bins = 10, alpha = 0.9, rwidth = 0.8, color = "purple", label = "PetalWidthCm")
plt.legend()
```

```python
plt.figure(figsize = (5,5))
s = data["SepalLengthCm"]
plt.hist(s, bins = 25, color = "brown", alpha = 0.7)
plt.title("SEPAL LENGTH IN CENTIMETER", fontsize = 16)
plt.xlabel("Sepal_Length_Cm", fontsize = 14)
plt.ylabel("Count", fontsize = 14)
```

```python
plt.figure(figsize = (5,5))
s = data["SepalWidthCm"]
plt.hist(s, bins = 25, color = "blue", alpha = 0.7)
plt.title("SEPAL WIDTH IN CENTIMETER", fontsize = 16)
plt.xlabel("Sepal_Width_Cm", fontsize = 14)
plt.ylabel("Count", fontsize = 14)
```

```python
plt.figure(figsize = (5,5))
s = data["PetalLengthCm"]
plt.hist(s, bins = 25, color = "red", alpha = 0.7)
plt.title("PETAL LENGTH IN CENTIMETER", fontsize = 16)
plt.xlabel("Petal_Length_Cm", fontsize = 14)
plt.ylabel("Count", fontsize = 14)
```

```python
plt.figure(figsize = (5,5))
s = data["PetalWidthCm"]
plt.hist(s, bins = 25, color = "orange", alpha = 0.7)
plt.title("PETAL WIDTH IN CENTIMETER", fontsize = 16)
plt.xlabel("Petal_Width_Cm", fontsize = 14)
plt.ylabel("Count", fontsize = 14)
```

# JointPlot Using Seaborn Library(ScatterPlot and Histogram)

```python
sns.jointplot(x = "SepalLengthCm", y = "PetalLengthCm", data = data, color = "red")
```

```python
sns.jointplot(x = "SepalWidthCm", y = "PetalWidthCm", data = data, color = "green")
```

```python
sns.jointplot(x = "SepalLengthCm", y = "PetalLengthCm", data = data, kind = "reg", color = "purple") 
```

```python
sns.jointplot(x = "SepalWidthCm", y = "PetalWidthCm", data = data, kind = "reg", color = "brown")
```
